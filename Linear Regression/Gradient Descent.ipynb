```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(X):
    return 3*X**2-4*X+5

def derFn(X):
    return 6*X-4
```

```python
X = np.linspace(-2.5,4,100)
y = fn(X)
```

```python
plt.plot(X,y)
plt.show()
```

```python
xm = 3
alpha = 0.01
xml = [xm]
```

```python
for i in range(200):
    xm = xm - alpha*derFn(xm)
    xml.append(xm)
```

```python
plt.plot(X,y)
plt.plot(xml,fn(np.array(xml)),'r')
plt.ylim(0,40)
plt.xlim(-3,5)
plt.show()
```
