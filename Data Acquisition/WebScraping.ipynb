```python
table_tags = soup.find_all('table')
```

```python
tb = table_tags[0]
```

```python
tb
```

```python
table_data = []
# Head
thead = tb.find('thead')
row = []
for th in thead.find_all('th'):
    row.append(th.text)
table_data.append(row)

#Row
tbody = tb.find('tbody')
for tr in tbody.find_all('tr'):
    # print(type(tr))
    row = []
    for td in tr.find_all('td'):
        row.append(td.text)

    table_data.append(row)


table_data
```

```python
table_data
```

```python
with open('Data.csv','w') as file:
    for row in table_data:
        file.write(','.join(row)+'\n')
```
